#***Data Frame***

```python
#Create DataFrame
data = {
    "Name":["A","B","c"],
    "Marks": [85,90,78]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Find The Structure Of The DataFrame
print(df.shape)
```

```python
#Size Of DataFrame
print(df.size)
```

```python
#Number Of Dimensions In A DataFrame
print(df.ndim)
```

```python
#Display Columns
print(df.columns)
```

```python
#Check DataType Of Each Column
print(df.dtypes)
```

```python
# Access Individual Column
print(df["Name"])
```

```python
# Access Name Of First Row
print(df.loc[1])
print(df.loc[0,"Name"])
print(df.iloc[2,1])
```

```python
#Add A Column
df["Grade"] = ["Firstclass","Distinction","Second Class"]
print(df)
```

```python
#Update Data In Dataframe
#Change Marks Of Student A to 88

df.iloc[0,1] =88
print(df)
```

```python
#Delete A Column
#df.drop("Grade",axis=1,inplace=True)
df1 =df.drop("Grade",axis=1)
print(df)
print(df1)
```

```python
#Perform Basic Statistical Analysis
mean1 = df["Marks"].mean()
print("Mean Value: ",mean1)
min1 = df["Marks"].min()
print("Min Value: ",min1)
max1 = df["Marks"].max()
print("Max Value: ",max1)
```

```python
#display student scoring more than 80
print(df[df["Marks"]>80])
```
